```python
from pathlib import Path
import pickle
import numpy as np
import pandas as pd
```

```python
import six.moves.cPickle


def load_dump(dump_file):
    dump = []
    with open(dump_file, "rb") as in_fd:
        while True:
            try:
                step = six.moves.cPickle.load(in_fd)
            except EOFError:
                return dump
            dump.append(step)


def trace_to_dataframe(trace: list[dict]) -> pd.DataFrame:
    """
    必要な要素だけ抽出して DataFrame 化
      step, reward, ball(x,y,z), scoreL, scoreR, action_num
    """
    rows = []
    for step, t in enumerate(trace):
        obs = t["observation"]
        dbg = t.get("debug", {})
        rows.append(
            {
                "step": step,
                "reward": t.get("reward", np.nan),
                "ball_x": obs["ball"][0],
                "ball_y": obs["ball"][1],
                "ball_z": obs["ball"][2],
                "score_L": obs["score"][0],
                "score_R": obs["score"][1],
            }
        )
    return pd.DataFrame(rows)
```
